```python
Y = df["Outcome"]
X_cont=df[df.columns[[0,1,2,3,4,5,6,7,]]]
X_cont 
X = X_cont
```

```python
from sklearn.preprocessing import StandardScaler
SS = StandardScaler()
SS_cont = SS.fit_transform(X_cont)
SS_cont = pd.DataFrame(SS_cont)
SS_cont.columns = list(X_cont)
SS_cont
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X,Y)
df["Y_pred"] = model.predict(X)
df["predict_proba"] = model.predict_proba(X)[:,1:]
```

```python
import streamlit as st
import numpy as np
intercept = -9.007
coefs = [0.064, 0.034, -0.014, 0.003, -0.002, 0.103, 0.627, 0.037]
feat_names = ["Pregnancies", "Glucose", "BloodPressure", "SkinThickness", "Insulin", "BMI", "DiabetesPedigreeFunction", "Age"]

st.title("Diabetes Prediction (Logistic Regression)")
st.write("Enter data below:")

X_input = []
for feat in feat_names:
    val = st.number_input(f"{feat}", value=0.0)
    X_input.append(val)

logit = intercept + np.dot(coefs, X_input)
prob = 1 / (1 + np.exp(-logit))

if st.button("Predict"):
    st.write(f"Predicted probability of diabetes: {prob:.2f}")
    st.write(f"{'Likely diabetic' if prob > 0.5 else 'Not likely diabetic'}")

st.info("This demo uses static model coefficients. To use your trained model, save it using joblib and load in Streamlit")
```
