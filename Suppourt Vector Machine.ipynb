```python
X_cont = df[df.columns[[0,1,2,3,4,5]]]
X_cont
```

```python
Y = df["Drug Response"]
X = X_cont.drop(["Drug Response"],axis=1)
X
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.20)
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import io
pairplot_fig = sns.pairplot(df, hue='Drug Response')
pairplot_fig.fig.suptitle('Feature Distributions and Relationships', y=1.02)
```

```python
plt.figure(figsize=(5,3))
sns.countplot(x='Drug Response', data=df)
plt.title('Class Distribution: Drug Response')
plt.xlabel('Drug Response')
plt.ylabel('Count')
```

```python
import numpy as np
from sklearn.svm import SVC
svc = SVC(kernel='linear',C=1)
svc.fit(X_train,Y_train)

Y_pred_train = svc.predict(X_train)
Y_pred_test = svc.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
score1 = (accuracy_score(Y_train , Y_pred_train))
score2 = (accuracy_score(Y_test , Y_pred_test))
print("Training Score:",np.round(score1,2))
print("Test Score:",np.round(score2,2))
```

```python
from sklearn.metrics import recall_score,f1_score,precision_score,confusion_matrix
score1 = (recall_score(Y_train , Y_pred_train))
score2 = (recall_score(Y_test , Y_pred_test))
print("Training Score for Recall:",np.round(score1,2))
print("Test Score for Recall:",np.round(score2,2))

score = (f1_score(Y_train , Y_pred_train))
score = (f1_score(Y_test , Y_pred_test))
print("Training Score for F1:",np.round(score,2))
print("Test Score for F1:",np.round(score,2))

prec_score = (precision_score(Y_train , Y_pred_train))
prec_score = (precision_score(Y_test , Y_pred_test))
print("Training Score for Precision:",np.round(prec_score,2))
print("Test Scorefor Precision:",np.round(prec_score,2))
```

```python
Y_pred_train = SVC(kernel='linear', random_state=42).fit(X_train, Y_train).predict(X_test)
```

```python
cm = confusion_matrix(Y_test, Y_pred_train)
fig, ax = plt.subplots(figsize=(4,4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Pred 0','Pred 1'], yticklabels=['Actual 0','Actual 1'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('SVM Confusion Matrix - Test Data')
plt.tight_layout()
```

```python
import numpy as np
from sklearn.svm import SVC
svc = SVC(kernel='poly',degree=3)
svc.fit(X_train,Y_train)

Y_pred_train = svc.predict(X_train)
Y_pred_test = svc.predict(X_test)

from sklearn.metrics import accuracy_score
score1 = (accuracy_score(Y_train , Y_pred_train))
score2 = (accuracy_score(Y_test , Y_pred_test))
print("Training Score:",np.round(score1,2))
print("Test Score:",np.round(score2,2))
```

```python
import numpy as np
from sklearn.svm import SVC
svc = SVC(kernel='rbf')
svc.fit(X_train,Y_train)

Y_pred_train = svc.predict(X_train)
Y_pred_test = svc.predict(X_test)

from sklearn.metrics import accuracy_score
score1 = (accuracy_score(Y_train , Y_pred_train))
score2 = (accuracy_score(Y_test , Y_pred_test))
print("Training Score:",np.round(score1,2))
print("Test Score:",np.round(score2,2))
```

Differences:
->| Kernel     | Accuracy | Precision | Recall | F1-score |
  | ---------- | -------- | --------- | ------ | -------- |
  | Linear     | 0.78     | 0.84      | 0.75   | 0.79     |
  | Polynomial | 0.69     | 0.86      | 0.54   | 0.66     |
  | RBF        | 0.82     | 0.87      | 0.80   | 0.83     |

-> By analyzing the data in different data visualizations the 'rbf' model is the suitable method for the data.

Strengths:
->Effective at modeling complex, nonlinear boundaries.
->Robust to outliers if proper regularization and kernel are chosen.
Weakness:
->Sensitive to feature scaling (all features should be normalized/standardized).
->
Kernel and regularization tuning required—can be computationally intensive with larger datasets or many hyperparameters.

Practical implimentations:
->SVM is excellent for real-world tasks where classes may not be linearly separable and where limited labeled data
exists like the medicaldata/pharma and fraud detection
