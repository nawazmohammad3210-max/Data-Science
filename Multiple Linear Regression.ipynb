```python
x_cont=df[df.columns[[0,1,2,4,5,6,7,8,9,10]]]
x_cont
```

```python
from sklearn.preprocessing import StandardScaler
SS = StandardScaler()
SS_cont = SS.fit_transform(x_cont)
SS_cont = pd.DataFrame(SS_cont)
SS_cont.columns = list(x_cont)
SS_cont
```

```python
Y = df['Price']
X = SS_cont.drop('Price',axis=1)
X
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.20)
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(X_train,Y_train)
```

```python
model.intercept_
```

```python
model.coef_
```

```python
Y_pred_train = model.predict(X_train)
Y_pred_test = model.predict(X_test)
```

```python
import numpy as np
from sklearn.metrics import mean_squared_error
test_MSE=(mean_squared_error(Y_test,Y_pred_test))
train_MSE=(mean_squared_error(Y_train,Y_pred_train))
print("Test MSE:",np.round(test_MSE,3))
print("Train MSE:",np.round(train_MSE,3))
```

```python
from sklearn.metrics import r2_score
print("Train R2 score:",np.round(model.score(X_train,Y_train),3))
print("Test R2 score:",np.round(model.score(X_test,Y_test),3))
```

```python
from sklearn.model_selection import ShuffleSplit,cross_validate
shuffle_split = ShuffleSplit(n_splits=500, test_size=0.2, random_state=42)
cv_scores = cross_validate(model,X, Y, cv=shuffle_split, scoring='r2',return_train_score=True)
train_scores = cv_scores['train_score']
test_scores = cv_scores['test_score']
print("cross validation: training score:" , np.round(train_scores.mean(),2))
print("cross validation: test score:" , np.round(test_scores.mean(),2))
```

```python
from sklearn.linear_model import LassoCV
from sklearn.model_selection import ShuffleSplit
import numpy as np
shuffle_split_lassocv = ShuffleSplit(n_splits=100, test_size=0.2, random_state=42)
lassocv_model = LassoCV(cv=shuffle_split_lassocv, random_state=42)
lassocv_model.fit(X, Y)
best_alpha_lassocv = lassocv_model.alpha_
mse_path = lassocv_model.mse_path_
mean_mse = mse_path.mean(axis=1)
best_mse = mean_mse[lassocv_model.alphas_ == best_alpha_lassocv][0]

print("\nLassoCV Results:")
print("Best alpha found:", best_alpha_lassocv)
print("Mean squared error for best alpha:", best_mse)
```

### Lasso

```python
from sklearn.linear_model import Lasso
final_lasso_model = Lasso(alpha=best_alpha_lassocv)
final_lasso_model.fit(X, Y)
lasso_coefficients = final_lasso_model.coef_
zero_coefficients = np.sum(lasso_coefficients == 0)
non_zero_coefficients = np.sum(lasso_coefficients != 0)

print("\nAnalysis of Lasso Coefficients (with best alpha):")
print("Number of zero coefficients:", zero_coefficients)
print("Number of non-zero coefficients:", non_zero_coefficients)
```

### Ridge

```python
from sklearn.linear_model import Ridge
Ridge_model = Ridge()
Ridge_model.fit(X_train,Y_train)
Ridge_Y_pred_train= Ridge_model.predict(X_train)
Ridge_Y_pred_test= Ridge_model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
train_Ridge_mse = mean_squared_error(Y_train,Ridge_Y_pred_train)
test_Ridge_mse = mean_squared_error(Y_test,Ridge_Y_pred_test)
print("Ridge Train MSE:",np.round(train_Ridge_mse,2))
print("Ridge Test MSE:",np.round(test_Ridge_mse,2))
```

```python
from sklearn.metrics import r2_score
print("Ridge Train R2 score:",np.round(Ridge_model.score(X_train,Y_train),2))
print("Ridge Test R2 score:",np.round(Ridge_model.score(X_test,Y_test),2))
```

NORMALIZATION:
Normalization means checking the data is it lies under 0 and 1.

Standardization:
Standardization is nothing but removing the units from the data and make sures thatunit free data.

it is helpful by the like the data contains no of datatypes which has units like kgs,cms,degrees,etc.


Centering variables:
Subtract the mean from correlated predictors to reduce multicollinearity caused by interaction or polynomial terms.
Variance Inflation Factor (VIF):
Calculate VIF for each variable to identify which predictors are highly collinear. Consider removing or transforming variables with high VIF.
