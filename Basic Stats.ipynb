```python
import numpy as np

numerical_cols = df.select_dtypes(include=np.number).columns

print("\n--- IQR Outlier Analysis for all Numerical Columns ---")
for col in numerical_cols:
    Q1 = np.percentile(df[col], 25)
    Q3 = np.percentile(df[col], 75)
    IQR = Q3 - Q1
    uw = Q3 + (1.5 * IQR)
    outliers_count = len(df[df[col] > uw])
    print(f"\n--- Column: {col} ---")
    lw = Q1 - (1.5 * IQR)
    outliers_count_lower = len(df[df[col] < lw])
    print(f"Total # of outliers: {outliers_count + outliers_count_lower}")
```
