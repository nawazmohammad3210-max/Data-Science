```python
from sklearn.metrics.pairwise import cosine_similarity
df = pd.read_csv('cleaned_anime.csv')
feature_matrix = pd.read_csv('anime_feature_matrix.csv')
cosine_sim = cosine_similarity(feature_matrix)
def get_recommendations(title, threshold=0.5, top_n=10):
    try:
        idx = df[df['name'].str.lower() == title.lower()].index[0]
    except IndexError:
        return "Anime not found."
    sim_scores = list(enumerate(cosine_sim[idx]))
    filtered_scores = [score for score in sim_scores if score[1] >= threshold and score[0] != idx]
    
    filtered_scores = sorted(filtered_scores, key=lambda x: x[1], reverse=True)
    top_scores = filtered_scores[:top_n]
    anime_indices = [i[0] for i in top_scores]

    result = df.iloc[anime_indices][['name', 'genre', 'type', 'rating']]
    result['similarity_score'] = [i[1] for i in top_scores]
    return result
test_anime = "Kimi no Na wa."
print(f"Recommendations for '{test_anime}' with threshold 0.5:")
print(get_recommendations(test_anime, threshold=0.5, top_n=5))

print(f"\nRecommendations for '{test_anime}' with strict threshold 0.9:")
print(get_recommendations(test_anime, threshold=0.9, top_n=5))
```

Collaborative Filtering (CF) is a recommendation method that relies on the "wisdom of the crowd." It predicts what you will like based on the behavior of other users, rather than the attributes of the item itself.
--->>User-Based: Matches People to People.
--->>Item-Based: Matches Items to Items.
--->>Collaborative Filtering: Uses interactions (ratings/clicks) instead of descriptions (genres/tags).
