```python
import numpy as np
from scipy.stats import t
data = [1.13, 1.55, 1.43, 0.92, 1.25, 1.36, 1.32, 0.85, 1.07, 1.48, 1.20, 1.33, 1.18, 1.22, 1.29]
n = len(data)
mean = np.mean(data)
std = np.std(data, ddof=1)
confidence = 0.99
alpha = 1 - confidence
df = n - 1
t_crit = t.ppf(1 - alpha/2, df)
sem = std / np.sqrt(n)
moe = t_crit * sem
lower = mean - moe
upper = mean + moe
{
    'mean': round(mean, 3),
    'std': round(std, 3),
    'n': n,
    'Confidence Level': f'{confidence*100:.0f}%',
    't_critical': round(t_crit, 3),
    'Margin of Error': round(moe, 3),
    'CI Lower': round(lower, 3),
    'CI Upper': round(upper, 3)
}
```

Steps :

->The population standard deviation is unknown.
The sample size is relatively small (n<30).
The t-distribution accounts for extra uncertainty introduced by estimating the population standard deviation from the sample.
->caluclate the standard mean error.
where s is stsndard devition 
n = smaple size
-> By cheking the CI values the intervel is lies between the CI values so the confidence  is 99%.

```python
import numpy as np
from scipy.stats import norm
pop_std = 0.2
sample_mean = 1.239
n = 15
confidence = 0.99
alpha = 1 - confidence
z_crit = norm.ppf(1 - alpha/2)
sem = pop_std / np.sqrt(n)
moe = z_crit * sem
lower = sample_mean - moe
upper = sample_mean + moe
{
    'mean': round(sample_mean, 3),
    'pop_std': pop_std,
    'n': n,
    'Confidence Level': f'{confidence*100:.0f}%',
    'z_critical': round(z_crit, 3),
    'Margin of Error': round(moe, 3),
    'CI Lower': round(lower, 3),
    'CI Upper': round(upper, 3)
}
```
