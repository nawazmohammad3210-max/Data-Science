Standardization:
->Standardization helps many algorithms converge faster and can improve performance where assumptions of normality are important. It doesn’t squash values into a bounded range, but makes their spread and center uniform.

Minmaxscaling:
->It keeps the values in the range of 0 to 1.
->Min-max scaling keeps all values proportionally spaced in a set range, so no variable dominates by magnitude.

Summary:
->Use normalization for distance or range-sensitive models and non-normal data.
->Use standardization for models assuming normality or needing centered, spread-out data.

```python
from sklearn.preprocessing import LabelEncoder
cat_cols = ['workclass', 'education', 'marital_status', 'occupation',
    'relationship', 'race', 'sex', 'native_country', 'income']
le = LabelEncoder()
for col in cat_cols:
    df[col + '_encoded'] = le.fit_transform(df[col])
print(df.head())
```

```python
bins = [0, 25, 40, 60, 100]
labels = ['<25', '25-40', '41-60', '60+']
df['age_group'] = pd.cut(df['age'], bins=bins, labels=labels, right=False)
df['age_group']
```

```python
#2. Map education to years
education_to_years = {'Preschool': 0, '1st-4th': 2, '5th-6th': 5, '7th-8th': 7, '9th': 9, '10th': 10,
    '11th': 11, '12th': 12, 'HS-grad': 12, 'Some-college': 13, 'Assoc-acdm': 14, 
    'Assoc-voc': 14, 'Bachelors': 16, 'Masters': 18, 'Doctorate': 20, 'Prof-school': 19}
df['education_num'] = df['education'].map(education_to_years)
df['education_num']
```

->Created bin-based age groups and mapped education levels to numeric years for better modeling and good understanding.
